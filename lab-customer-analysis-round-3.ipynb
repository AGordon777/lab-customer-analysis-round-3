```python
import pandas as pd
import numpy as np

customer = pd.read_csv('marketing_customer_analysis.csv')

#csv and excel were identical, so only csv read
```

```python
# Task 1: Show DataFrame info
customer #returns the dataset
```

```python
# Task 2: Describe the data
customer.describe(include='all').T # transposed
#can see some numeric and some category data
```

```python
# Task 3: Show a plot of the total number of responses
import matplotlib.pyplot as plt #this is a go-to plotting tool for python

#choose a barplot as this is good for representing categorised data
plt.figure(figsize=(8, 6))
customer['Response'].value_counts().plot(kind='bar')
plt.xlabel('Responses')
plt.ylabel('Count')
plt.title('Total Number of Responses')
plt.show()
```

```python
# Task 4: Show a plot of the response rate by the total claim amount.

# Convert 'Response' column to 0s and 1s to generate a RATE
customer['Response'] = customer['Response'].map({'No': 0, 'Yes': 1})

#example list = ['Yes','No','No','No']
#converted list = [1,0,0,0]
#mean(list) = 0.25


# Plot the response rate by the sales channel by grouping

response_rate_by_channel = customer.groupby('Sales Channel')['Response'].mean()

#this groups the data by one variable 'Sales Channel' and then calculated the mean of another variable 'Response' which we changed to 1 for Response=Yes and 0 for Response=No
#we can only do this as the Response column has been replaced by 1s and 0s


response_rate_by_channel
```

```python
plt.figure(figsize=(8, 6))
response_rate_by_channel.plot(kind='bar')
plt.xlabel('Sales Channel')
plt.ylabel('Response Rate')
plt.title('Response Rate by Sales Channel')
plt.show()
```

```python
# Task 5: Show a plot of the response rate by the total claim amount.

#we need to classify customers into total claim amount bins - maybe using quartiles?

claim_q25 = np.quantile(customer['Total Claim Amount'], q=0.25)
claim_q50 = np.quantile(customer['Total Claim Amount'], q=0.50)
claim_q75 = np.quantile(customer['Total Claim Amount'], q=0.75)

#how do I group these? use a CUT function to add a new column with categories

bins = [0,claim_q25,claim_q50,claim_q75,float('inf')] # one more element here than for labels as we define the LIMITS
labels = ['1st quartile','2nd quartile','3rd quartile','4th quartile']

customer['Claim Amount Category']=pd.cut(customer['Total Claim Amount'], bins=bins, labels=labels, right=False)

customer
```

```python
#again calculate a grouped mean value for Responses

response_rate_by_claim_amount = customer.groupby('Claim Amount Category')['Response'].mean()

response_rate_by_claim_amount
```

```python
plt.figure(figsize=(8, 6))
response_rate_by_claim_amount.plot(kind='bar')
plt.xlabel('Claim Amount Quartile')
plt.ylabel('Response Rate')
plt.title('Response Rate by Claim Amount')
plt.show()
```

```python
# Task 6: Show a plot of the response rate by Income.

income_q25 = np.quantile(customer['Income'], q=0.25)
income_q50 = np.quantile(customer['Income'], q=0.50)
income_q75 = np.quantile(customer['Income'], q=0.75)

# Try grouping by quartiles

# bins1 = [0,income_q25,income_q50,income_q75,float('inf')]
# labels1 = ['1st quartile','2nd quartile','3rd quartile','4th quartile']

# customer['Income Category']=pd.cut(customer['Income'], bins=bins1, labels=labels1, right=False)
# income_q25

#BUT it errors as income_q25 =0

#So group just into low and high income (above / below median)

bins1 = [0,income_q50,float('inf')]
labels1 = ['low income','high income']

customer['Income Category']=pd.cut(customer['Income'], bins=bins1, labels=labels1, right=False)

customer
```

```python
response_rate_by_income = customer.groupby('Income Category')['Response'].mean()

response_rate_by_income
```

```python
plt.figure(figsize=(8, 6))
response_rate_by_income.plot(kind='bar')
plt.xlabel('Income Category')
plt.ylabel('Response Rate')
plt.title('Response Rate by Income')
plt.show()

#not very interesting
```
